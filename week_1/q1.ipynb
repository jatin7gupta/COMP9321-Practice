Activity-1: CSV Files
Description : You will learn how to into CSV files pandas' dataframe and how to create a CSV file out of a dataframe

Steps :

Download the dataset
Read the CSV file and put the samples into a pandas' dataframe.
Programmatically print the columns of the dataframe
Programmatically print the rows of the dataframe
Save the dataframe as a CSV file


```python
#importing libary
import pandas as pd
```

```python
#getting data
df = pd.read_csv('./data/Demographic_Statistics_By_Zip_Code.csv')
df
```

```python
#printing columns
for i in df:
    print(i)
```

```python
#print rows
count_row = df.shape[0]
for i in range(0, count_row):
    print(df.iloc[i])
    print()
```

```python
#save data 
df.to_csv('./data/export.csv', index=None, header=True)
```

```python
#solution
def print_dataframe(dataframe, print_column=True, print_rows=True):
    # print column names
    if print_column:
        print(",".join([column for column in dataframe]))

    # print rows one by one
    if print_rows:
        for index, row in dataframe.iterrows():
            print(",".join([str(row[column]) for column in dataframe]))

print_dataframe(df)
```

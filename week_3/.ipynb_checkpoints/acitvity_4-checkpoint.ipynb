```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/iris.csv')
df.groupby('species').count()
df_setosa = df.query('species=="setosa"')
df_virginica = df.query('species=="virginica"')
df_vercicolor = df.query('species=="versicolor"')

fig, axes = plt.subplots(nrows=1, ncols=2)

ax = df_setosa.plot.scatter(x='sepal_length', y='sepal_width', label='setosa', ax=axes[0])
ax = df_vercicolor.plot.scatter(x='sepal_length', y='sepal_width', label='versicolor', color='green', ax=ax)
ax = df_virginica.plot.scatter(x='sepal_length', y='sepal_width', label='virginica', color='red', ax=ax)

ax = df_setosa.plot.scatter(x='petal_length', y='petal_width', label='setosa', ax=axes[1])
ax = df_vercicolor.plot.scatter(x='petal_length', y='petal_width', label='versicolor', color='green', ax=ax)
ax = df_virginica.plot.scatter(x='petal_length', y='petal_width', label='virginica', color='red', ax=ax)
plt.show()
```
